```python
import pandas as pd
import numpy as np
```

Import the csv, do not skip blanks

```python
input = pd.read_csv("input.txt", sep = " ",header= None, names= ["opponent","response"])
df = pd.DataFrame(input)
```

"The first column is what your opponent is going to play: A for Rock, B for Paper, and C for Scissors. The second column--" Suddenly, the Elf is called away to help with someone's tent.

The second column, you reason, must be what you should play in response: X for Rock, Y for Paper, and Z for Scissors. Winning every time would be suspicious, so the responses must have been carefully chosen.

The winner of the whole tournament is the player with the highest score. Your total score is the sum of your scores for each round. 

Defining the conditions for the score of shape selection inside a function

```python
def shape_score(shape):
    if shape == 'X':
        return 1
    elif shape == 'Y':
        return 2
    elif shape == 'Z':
        return 3
```

Assigning score per row for selection

```python
df['selectedscore'] = df['response'].apply(shape_score) 
df
```

Mapping opponent and response actions in a dictionary

```python
opponent_dict = {  "A": "rock",  "B": "paper",  "C": "scissors"}
response_dict = {  "X": "rock",  "Y": "paper",  "Z": "scissors"}
```

Replace response by rock, paper, scissors.

```python
df
```

```python
df['opponent_action'] = df['opponent'].map(opponent_dict)
df['response_action'] = df['response'].map(response_dict)
df
```

Function with the game conditions. The score for a single round is the score for the shape you selected (1 for Rock, 2 for Paper, and 3 for Scissors) plus the score for the outcome of the round (0 if you lost, 3 if the round was a draw, and 6 if you won). 

```python
def rps_score(opponent, response):
    if response == opponent:
        return 3
    elif response == "rock":
        if opponent == "scissors":
            return 6
        else:
            return 0
    elif response == "paper":
        if opponent == "rock":
            return 6
        else:
            return 0
    elif response == "scissors":
        if opponent == "paper":
            return 6
        else:
            return 0
```

Assigning score per row with win conditions

```python
df['rps_score']= df.apply(lambda x: rps_score(x['opponent_action'], x['response_action']), axis=1)
df
```

Output 1: Sum of scores

```python
result1 = np.sum(df['selectedscore']) + np.sum(df['rps_score'])
print(result1)
```

Output 2: 

"Anyway, the second column says how the round needs to end: X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win. Good luck!"

```python
df
```

```python
response_result_dict = {  "X": "loss",  "Y": "draw",  "Z": "win"}
df['result'] = df['response'].map(response_result_dict)
df
```

LOSE:
opp: rock > scissors
opp: scissors > paper
opp: paper > rock

WIN :
opp: rock > paper
opp: scissors > rock
opp: paper > scissors

draw :
opp: rock > rock
opp: scissors > scissors
opp: paper > paper

```python
def response_action(opponent_action, result):
    if result == "draw":
        return opponent_action
    elif result == "loss":
        if opponent_action == "rock":
            return "scissors"
        elif opponent_action == "scissors":
            return "paper"
        else:
            return "rock"
    elif result == "win":
        if opponent_action == "rock":
            return "paper"
        elif opponent_action == "scissors":
            return "rock"
        else:
            return "scissors"
```

```python
df['response_action']= df.apply(lambda x: response_action(x['opponent_action'], x['result']), axis=1)
df
```

```python
def shape_score2(shape):
    if shape == 'rock':
        return 1
    elif shape == 'paper':
        return 2
    elif shape == 'scissors':
        return 3
```

```python
df['rps_score2']= df.apply(lambda x: rps_score(x['opponent_action'], x['response_action']), axis=1)
df
```
